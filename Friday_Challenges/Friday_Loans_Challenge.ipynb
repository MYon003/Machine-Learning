```python
import pandas as pd

# Putting the train data into a table
loan_train = pd.read_csv('https://raw.githubusercontent.com/zacharski/ml-class/master/data/loan_train.csv', na_values='-')
loan_train = loan_train.dropna()
loan_train = loan_train.set_index('Loan_ID')
loan_train[:10]
```

```python
# Make troublesome columns into usable data for the classifier
loan_hot = pd.get_dummies(loan_train[['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Dependents']])
loan_train = loan_train.drop('Gender', axis=1)
loan_train = loan_train.drop('Married', axis=1)
loan_train = loan_train.drop('Education', axis=1)
loan_train = loan_train.drop('Self_Employed', axis=1)
loan_train = loan_train.drop('Property_Area', axis=1)
loan_train = loan_train.drop('Dependents', axis=1)
loan_hot[:10]
```

```python
loan_train = loan_train.join(loan_hot)
loan_train[:10]
```

```python
# Putting the test data into a table
loan_test = pd.read_csv('https://raw.githubusercontent.com/zacharski/ml-class/master/data/loan_test.csv',
                     na_values='-')
loan_test = loan_test.set_index('Loan_ID')
loan_test = loan_test.dropna()
loan_test = loan_test.drop(columns = 'Unnamed: 0')
loan_test[:10]
```

```python
# Make troublesome columns into usable data for the classifier
loan_hot2 = pd.get_dummies(loan_test[['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Dependents']])
loan_test = loan_test.drop('Gender', axis=1)
loan_test = loan_test.drop('Married', axis=1)
loan_test = loan_test.drop('Education', axis=1)
loan_test = loan_test.drop('Self_Employed', axis=1)
loan_test = loan_test.drop('Property_Area', axis=1)
loan_test = loan_test.drop('Dependents', axis=1)
loan_hot2[:10]
```

```python
loan_test = loan_test.join(loan_hot2)
loan_test[:10]
```

```python
# Get the Features and Labels
featureNames = list(loan_train.columns)
featureNames.remove('Loan_Status')
print(featureNames)
```

```python
loan_train_features = loan_train[featureNames]
loan_train_labels = loan_train['Loan_Status']

loan_test_features = loan_test[featureNames]
```

```python
# Make the classifier
from sklearn import tree
from sklearn.ensemble import BaggingClassifier

clf = tree.DecisionTreeClassifier(criterion='entropy')

loan_clf = BaggingClassifier(clf, n_estimators=100, max_samples=.7, 
                                bootstrap=True, n_jobs=-1)

loan_clf.fit(loan_train_features, loan_train_labels)
predictions = loan_clf.predict(loan_test_features)
print(predictions)
```

```python
# Turning the predictions into nums
with open('predictions.txt', 'w') as f:
  for x in predictions:
    if x == 'Y':
      f.write("1\n")
    elif x == 'N':
      f.write("0\n")
```
